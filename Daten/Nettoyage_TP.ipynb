```python
import pandas as pd

# Charger le fichier CSV dans un DataFrame
#Remplacer le nom par ficher voulu
nom_fichier = '50TP-TA_Presidentielle_2022__1er janvier au 7 mars 2022_v2.csv'  # Remplacez 'votre_fichier.csv' par le nom de votre fichier
encodage = 'ISO-8859-1'  # Essayez également 'latin1' si cela ne fonctionne pas

df = pd.read_csv(nom_fichier, encoding=encodage)

print("\nInformations sur le DataFrame :")
print(df.info())

print("\nStatistiques descriptives du DataFrame :")
print(df.describe())

```

```python


# Charger le fichier CSV en spécifiant l'encodage
df = pd.read_csv(nom_fichier, encoding=encodage, delimiter=';')

# Filtrer les lignes où la colonne "Pourcentage" n'est pas vide ou contient un tiret '-'
df = df[(df['Pourcentage'].notna()) & (df['Pourcentage'] != '-')]

# Afficher les premières lignes du DataFrame
print("Aperçu des premières lignes du DataFrame :")
print(df.head())

# Afficher des informations sur le DataFrame
print("\nInformations sur le DataFrame :")
print(df.info())

# Afficher des statistiques descriptives
print("\nStatistiques descriptives du DataFrame :")
print(df.describe())
```
